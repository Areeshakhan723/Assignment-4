# **Problem Statement**
Guess My Number

I am thinking of a number between 0 and 99... Enter a guess: 50 Your guess is too high

Enter a new number: 25 Your guess is too low

Enter a new number: 40 Your guess is too low

Enter a new number: 45 Your guess is too low

Enter a new number: 48 Congrats! The number was: 48

```python
import random

def main():
    guess_num = random.randint(0, 99)  # The number to guess
    print("I am thinking of a number between 0 and 99...")

    user_input = int(input("Enter a number to guess: "))  # Get first guess

    while user_input != guess_num:
        if user_input < guess_num:
            print(f"Your guess is too low: {user_input}")
        else:
            print(f"Your guess is too high: {user_input}")

        # Ask for the next guess
        user_input = int(input("Enter a new number to guess: "))

    print(f"Congrats! The number was: {guess_num}")

if __name__ == '__main__':
    main()
```
