```python
def is_odd(num : int):
  """
    Checks to see if a value is odd. If it is, returns true.
  """
  remainder = num % 2
  return remainder == 1


def main():
  for i in range(10):
    if is_odd(i):
      print(f"Odd {i}")
    else:
      print(f"Even {i}")


if __name__ == '__main__':
  main()
```
