# **Problem Statement**
There's a small fruit shop nearby your house that you like to buy from. Since you buy several fruit at a time, you want to keep track of how much the fruit will cost before you go. Luckily you wrote down what fruits were available and how much one of each fruit costs.

Write a program that loops through a dictionary of fruits, prompting the user to see how many of each fruit they want to buy, and then prints out the total combined cost of all of the fruits.

Here is an example run of the program:

How many (apple) do you want?: 2

How many (durian) do you want?: 0

How many (jackfruit) do you want?: 1

How many (kiwi) do you want?: 0

How many (rambutan) do you want?: 1

How many (mango) do you want?: 3

Your total is $99.5

```python
def main():

   # Dictionary storing fruit names and their prices
  fruits = {
    "banana": 2,
    "grape": 45,
    "pineapple": 75,
    "kiwi": 1.2,
    "rambutan": 1.5,
    "mango": 5,
    "strawberry": 3,
    "blueberry": 4,
    "watermelon": 10
  }

  while True:

    totalPrice = 0 # Reset total price for each order

    # Loop through each fruit in the dictionary
    for fruit,price in fruits.items():

      while True:  # Keep asking until a valid number is entered

        try:

          userInput = input(f"How many {fruit}s do you want?: ")

          if userInput == "": # If empty, assume 0
            userInput = 0
          else:
           userInput = int(userInput) # Convert only if input is not empty


          if userInput < 0: # Prevent negative values
            print("Please enter a valid number (0 or more).")
            continue


          totalPrice += price * userInput # Calculate total cost
          break  # Exit loop when valid input is provided

        except ValueError:
          print("Invalid input! Please enter a valid number.")
          continue

    # Display the total price of the order
    print(f"\nðŸ›’ Your total is **${totalPrice:.2f}**")

    more = input("Would you like to order again? (yes/no): ").strip().lower()
    if more != "yes": # Exit if the user says no
      print("âœ… Thank you for shopping with us! Have a great day. ðŸ˜Š")
      break



if __name__ == '__main__':
    main()
```
