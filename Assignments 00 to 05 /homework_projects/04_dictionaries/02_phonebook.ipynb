# **Problem Statement**
In this program we show an example of using dictionaries to keep track of information in a phonebook.

```python
# function read_phone_numbe()
def read_phone_number():
  """
    This function collects user names and phone numbers and stores them in a dictionary.
    The user can enter multiple contacts, and an empty name input stops the process.
  """
  phoneBook = {}  # create a empty dictionary to store user data

  while True:
    userName = input("Enter your name: ")

    if userName == "":  # Stop if the user enters an empty string
      break

    while True:  # Keep asking for a valid number

      userNumber = input("Enter your number: ")
      if userNumber.strip():  # Ensure it's not empty (strip removes spaces)
        break
      print("please enter a valid phone number")  # Warn user about empty input


    phoneBook[userName] = userNumber # Store the name-number pair in the dictionary

  return phoneBook  # Return the completed phonebook




# function of print_phone_number()
def print_phone_number(phonebook):
  """
    Prints out all the names/numbers in the phonebook.
  """

  for name, number in phonebook.items():  # Loop through each entry in the dictionary
    print(f"{name} -> {number}")



# function of lookup_numbers()
def lookup_numbers(phonebook):
   """
    Allows the user to look up phone numbers by entering a name.
    If the name exists in the phonebook, the number is displayed.
    Otherwise, a message is shown that the name is not in the phonebook.
   """
   while True:
    userLookup = input("Enter the name to look up: ")  # Ask for a name to search

    if userLookup == "":  # Stop lookup process if the input is empty
      break


    if userLookup not in phonebook:  # Check if the name is not in the phonebook
      print(f"{userLookup} is not in the phonebook")  # Inform user
    else:
        print(f"{userLookup} -> {phonebook[userLookup]}")  # Display the number




# main function()
def main():
  """
  This function runs the complete program:
    - Reads phone numbers from user input
    - Prints the stored phonebook
    - Allows the user to look up contacts
  """
  phonebook = read_phone_number() # Collect user input and create phonebook
  print_phone_number(phonebook) # Display the phonebook
  lookup_numbers(phonebook) # Allow the user to look up contacts


if __name__ == '__main__':
    main()
```
