# **Problem Statement**
Converts feet to inches. Feet is an American unit of measurement. There are 12 inches per foot. Foot is the singular, and feet is the plural.



```python
def main():
  """
    This program converts a user-provided length in feet to inches
    using the standard conversion factor (1 foot = 12 inches).
  """

  inches_per_foot = 12  # Conversion factor: 1 foot = 12 inches

  feet = float(input("Enter numbers of feet: "))  # user inpur feet
  inches = feet * inches_per_foot  # convert feet to inches
  print(f"{feet} feet is equal to {inches} inches!")


if __name__ == "__main__":
  main()
```
