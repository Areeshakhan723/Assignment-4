# **Problem Statement**
Write a program which prompts the user for an adjective, then a noun, then a verb, and then prints a fun sentence with those words!

Mad Libs is a word game where players are prompted for one word at a time, and the words are eventually filled into the blanks of a word template to make an entertaining story! We've provided you with the beginning of a sentence (the SENTENCE_START constant) which will end in a user-inputted adjective, noun, and then verb.

Here's a sample run

Please type an adjective and press enter. tiny

Please type a noun and press enter. plant

Please type a verb and press enter. fly

Code in Place is fun. I learned to program and used Python to make my tiny plant fly!

```python
def main():
  """
  Mad libs game where the user provide word to complete a funy story!
  """

  print("Welcome to Mad Libs! Fill in the blanks to create a hilarious story.\n")

  # Get user inputs
  name = input("Enter a name: ")
  animal = input("Enter an animal: ")
  place = input("Enter a place: ")
  adjective = input("Enter an adjective (describes something): ")
  verb = input("Enter a verb (action word): ")
  food = input("Enter a type of food: ")
  exclamation = input("Enter an exclamation (e.g., Wow!, Yikes!, Oh no!): ")

  # Create the funny story
  story = f"""
  One day, {name} went to {place} and saw a {adjective} {animal}.
  "{exclamation}!" {name} shouted. The {animal} started to {verb} wildly!
  In a panic, {name} threw a {food} at it, hoping it would calm down.
  Instead, the {animal} ate the {food} and became even more {adjective}!
  Moral of the story: Never {verb} at a {animal} while holding {food} in {place}.
  """

  # Print the completed story
  print("\nHere’s your hilarious Mad Libs story:\n")
  print(story)

if __name__ == "__main__":
    main()
```
