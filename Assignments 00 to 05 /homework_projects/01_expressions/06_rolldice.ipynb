# **Problem Statement**
Simulate rolling two dice, and prints results of each roll as well as the total.

```python
import random # import the random module for generate random numbers

# Define the number of side on a rolling dice (standard dice has 6 sides)
rolling_dice = 6

def main():
  # Generate a individual number between 1 to 6 for each side of dice roll
  dice1 = random.randint(1, rolling_dice)
  dice2 = random.randint(1, rolling_dice)


  # Display the individual dice result
  print(f"dice 1: {dice1}")
  print(f"dice 2: {dice2}")

  # Calculate the both dice
  total = dice1 + dice2

  # Display the total result od both dice
  print(f"Total of Two dice: {total}")


# call the main function
if __name__ =="__main__":
  main()
```
