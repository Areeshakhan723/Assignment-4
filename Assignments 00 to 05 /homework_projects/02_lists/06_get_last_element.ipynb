# **Problem Statement**
Fill out the function get_last_element(lst) which takes in a list lst as a parameter and prints the last element in the list. The list is guaranteed to be non-empty, but there are no guarantees on its length.

```python
def get_last_element(user_list):
  """
  This function take a list as a parameter and prints the last element in the list.
  """

  if user_list:
    print(f"Last element: {user_list[-1]}")
  else:
    print("The list is empty!")


def list_of_user():
  userInput = input("Enter your favorite things (separated by spaces): ").split()

  if userInput:
    return userInput
  else:
    print("Enter atleast one element in the list!")
    return []


def main():
  lst = list_of_user()
  get_last_element(lst)


if __name__ == '__main__':
    main()
```
