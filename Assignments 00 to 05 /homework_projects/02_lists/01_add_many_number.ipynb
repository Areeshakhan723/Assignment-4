# **Problem Statement**
Write a function that takes a list of numbers and returns the sum of those numbers.

```python
# Function to calculate the sum of the list of numbers
def sum_of_numbers(numbers):
  """
  This program takes list of numbers from the user and returns the same of those numbers!
  """
  total = 0

  for nums in numbers: # Iterate through each number in the list
    total += nums   # Add the number to the total sum
  return total  # return the total value of sum


# Function to take the user input and display the result
def add_many_numbers():
  # Prompt user to enter numbers separated by spaces, then convert them into a list of integers
  userInput = list(map(int,input("Enter numbers by separated a space: ").split()))

  # Call sum_of_numbers function to compute the sum of user input
  total =  sum_of_numbers(userInput) # Call the sum_of_numbers function

  print(f"Your Input: {userInput}") # Display the list of user number
  print(f"Total numbers of sum: {total}") # Display the total sum of numbers


if __name__ == "__main__":
  add_many_numbers()
```
