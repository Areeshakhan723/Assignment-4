# **Problem Statement**
Write a program which continuously asks the user to enter values which are added one by one into a list. When the user presses enter without typing anything, print the list.

Here's a sample run (user input is in blue):

Enter a value: 1 Enter a value: 2 Enter a value: 3 Enter a value: Here's the list: ['1', '2', '3']

```python
def main():

  """
  This program continuously asks the user enter values which are added one by one into a list.
  """

  user_list = [] # create a empty list

  userInput = input("Enter a value: ") # ask the user's to enter the value

  while userInput:

      user_list.append(userInput) # append the list
      userInput = input("Enter a value: ") # ask the user's to enter the value

  print(f"Here's is list: {user_list}") # print list



if __name__ =="__main__":
  main()
```
