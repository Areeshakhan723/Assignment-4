# **Problem Statement**
Fill out the function get_first_element(lst) which takes in a list lst as a parameter and prints the first element in the list. The list is guaranteed to be non-empty. We've written some code for you which prompts the user to input the list one element at a time.

```python
def get_first_element(first_element):
  """
    Prints the first element of a provided list.
  """

  if first_element: # Check if list is not empty
    print(first_element[0])
  else:
    print("The list is empty!")



def list_of_user():
  """
    Prompts the user to enter one element of the list at a time and returns the resulting list.
  """

  userInput = input("Enter a list :").split()

  if userInput: # If user entered something
    return userInput
  else:
    print("Please enter at least one element!")
    return[]  # Return empty list if no input

def main():
    # Get list from user
    lst = list_of_user()

    # Print the first element
    get_first_element(lst)



if __name__ == '__main__':
    main()
```
