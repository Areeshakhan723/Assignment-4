# **Problem Statement**
Write a program that doubles each element in a list of numbers. For example, if you start with this list:

numbers = [1, 2, 3, 4]

You should end with this list:

numbers = [2, 4, 6, 8]

```python
def main():


  numlist: list = [1,2,3,4] # create  a list of numbers


  for i in range(len(numlist)):  # Loop through the indices of the list
    element_of_index = numlist[i] # Get the element at index i in the numbers list
    numlist[i] = element_of_index * 2


  print(numlist)  # This should print the doubled list



if __name__ == "__main__":
  main()
```
