# **Problem Statement**
Fill out the function shorten(lst) which removes elements from the end of lst, which is a list, and prints each item it removes until lst is MAX_LENGTH items long. If lst is already shorter than MAX_LENGTH you should leave it unchanged. We've written a main() function for you which gets a list and passes it into your function once you run the program. For the autograder to pass you will need MAX_LENGTH to be 3, but feel free to change it around to test your program.

```python
MAX_LENGTH: int = 3  # Maximum allowed length of the list

def shorten(lst):
  """Removes extra elements until the list length is MAX_LENGTH."""
  while len(lst) > MAX_LENGTH: # Keep removing elements if the list is too long
    removed_element = lst.pop()   # Remove last element
    print(f"Removed element: {removed_element}")
    print(f"Shortened list: {lst}")

def get_list():

  """Collects user input into a list until the user presses Enter."""

  my_list = [] # create a empty list
  print("Enter numbers one by one (press Enter to finish): ")

  while True:
    userInput = input("Enter a list of numbers: ").strip()

    if userInput == "":  # Stop when input is empty
      break
    my_list.append(userInput)  # Append input to the list

  return my_list


def main():
  lst = get_list() # Get list from user
  print(f"Original list: {lst}")  # Show original list

  shorten(lst) # Shorten list

  print(f"Final list (max {MAX_LENGTH} elements): {lst}")  # Show final list

if __name__ == '__main__':
    main()
```
