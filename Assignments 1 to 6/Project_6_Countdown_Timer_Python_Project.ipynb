# **Project 5. Countdown Timer**

```python
import time

def countdown(t):
  while t > 0:
    mins, secs = divmod(t, 60)
    timer = f"{mins:02d} : {secs:02d}"
    print(timer, flush=True)
    time.sleep(1)
    t -= 1

  print("\nTimer completed!")

try:
  t = input("Enter a time in seconds! ")
  print(f"Starting countdown from {t} seconds...")
  countdown(int(t))
except ValueError:
  print("Please enter a valid number!")
```
