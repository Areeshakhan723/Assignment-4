# **Project 3. Guess the Number (user)**

```python
import random

"""
 This program lets the computer guess a number that the user is thinking of.
 The user gives feedback if the guess is too high, too low, or correct.
 The computer keeps adjusting its guesses until it finds the right number.
"""
def main(x):

  # Set the initial low and high values for the guessing range
  low = 1
  high = x
  feedback = ""  # Variable to store user's feedback on the guess


 # Loop until the user indicates the guess is correct
  while feedback != "c":
    if low != high:
      guess = random.randint(low, high) # Generate a guess within the range
    else:
      guess = low  # could also be high becouse low = high


   # Ask for user feedback on whether the guess is too high, too low, or correct
    feedback = input(f"Is {guess} too high (H), too low (L), correct (C)!").lower()


    # Adjust the high or low range based on feedback
    if feedback == "h":
      high = guess - 1  # The guess was too high, so we adjust the high boundary
    elif feedback == "l":
      low = guess + 1   # The guess was too low, so we adjust the low boundary


  # Once the correct guess is made, print the success message
  print(f"Yay! The computer gussed your number {guess} correctly!")


if __name__ == '__main__':
    main(10)
```
