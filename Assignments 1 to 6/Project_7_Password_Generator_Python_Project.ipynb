# **Project 7. Password_Generator**

```python
import random
import string

char = string.ascii_letters + string.digits + string.punctuation

numbers = int(input("Enter a amount of password generate: ")) # ask the user amount of password

length = int(input("Enter a length of password you want: ")) # ask the user length of password character

print("\nHere are your passwords")

for pwd in range(numbers):
  password = ""
  for c in range(length):
    password += random.choice(char)
  print(password)
```
