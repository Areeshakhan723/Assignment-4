# **Project 4. Rock paper scissors**
The program is a Rock, Paper, Scissors game where the user inputs a choice, and the computer randomly selects one. If both choices match, it's a tie. Otherwise, the is_win() function checks if the user wins based on standard game rules. The result is then displayed.











```python
import random

def play():
  user = input("What't your choice? 'R' for rock, 'P' for paper, 'S' for scissors").lower()
  computer = random.choice(["r", "p", "s"])

  print(f"Computer chose: {computer}")
  print(f"You choose: {user}")

  if user == computer: # If both choices match, it's a tie.
    return "It's tie"

  # r beats s, s beats p, p beats r
  if is_win(user, computer):
    return "You win!"

  return "You lost!"



def is_win(player, opponent):
   # r beats s, s beats p, p beats r
  if (player == "r" and opponent == "s") or \
     (player == "s" and opponent == "p") or \
     (player == "p" and opponent == "r"):
     return True

  return  False  # Added return for losing cases


print(play())
```
