```python
import os
import json
import shutil
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the folder path where your files are stored
folder_path = "/content/drive/My Drive/PROJECTS/Assignments 01/02_intermediate/Intro to Python"
os.makedirs(folder_path, exist_ok=True)  # Ensure the folder exists

# List of Jupyter Notebook files
file_names = [
  "Planetary Weights Solution.ipynb"
]

# Define the content to add
notebook_content = {
    "cells": [
        {
            "cell_type": "markdown",
            "metadata": {},
            "source": [
                "# This is a Text Cell\n",
                "You can write your assignment description here.\n"
            ]
        },
        {
            "cell_type": "code",
            "execution_count": None,
            "metadata": {},
            "outputs": [],
            "source": [
                "# This is a Code Cell\n",
                "print('Hello, Colab!')\n"
            ]
        }
    ],
    "metadata": {},
    "nbformat": 4,
    "nbformat_minor": 4
}

# Loop through all files and add content
for file in file_names:
    file_path = os.path.join(folder_path, file)

    try:
        # Check if the file exists
        if os.path.exists(file_path):
            with open(file_path, "r", encoding="utf-8") as f:
                notebook_data = json.load(f)
                notebook_data["cells"].extend(notebook_content["cells"])  # Append new cells

            # Backup the existing file before modification
            shutil.copy(file_path, file_path + ".bak")

        else:
            notebook_data = notebook_content  # If file doesn't exist, create it

        # Save the updated notebook
        with open(file_path, "w", encoding="utf-8") as f:
            json.dump(notebook_data, f, indent=2)

        print(f"✅ Cells added to: {file}")

    except (json.JSONDecodeError, FileNotFoundError) as e:
        print(f"⚠️ Error processing {file}: {e}")
```
